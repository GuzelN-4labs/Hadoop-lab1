### Добавлю скользящую среднюю с окном в 30 дней для курса закрытия ('Closed')

```python
df['moving_avg_30_D'] = df.sort_values(by=['Date'], ascending = True)['Close'].rolling(window = 30, min_periods = 1).mean()
```

```python
df[24:34]
```

```python
plt.figure(figsize = (10,7))
plt.plot(df['Date'],df['Close'], color = "orange", label = "Close")
plt.plot(df['Date'], df['moving_avg_30_D'], color = "darkgreen", label = "moving avg (30 days)")
plt.title('Apple stocks 2')
plt.xlabel('Dates')
plt.ylabel('Rates')
plt.legend()
plt.show()
```

```python
df['open_close_diff'] = df['Close'] - df['Open']
df['day_of_week'] = df['Date'].dt.dayofweek + 1
dayofweek_names = dict({1: "Monday",
                         2: "Tuesday",
                         3: "Wednesday" ,
                         4: "Thursday",
                         5: "Friday",
                         6: "Saturday",
                         7: "Sunday"})
df['Day_names'] = df['day_of_week'].map(dayofweek_names)
df = df.sort_values(by = 'Date', ascending = True)
```

```python
df.groupby('Day_names')['open_close_diff'].mean().reset_index().plot('Day_names', 'open_close_diff',
                      subplots=True,
                      kind="line",
                      title = (f"Courses diff"),
                      color = "forestgreen",
                      label = f"Courses diff" )
```

```python
df.groupby('Day_names')['Volume'].mean().reset_index().plot('Day_names', 'Volume',
                      subplots=True,
                      kind="line",
                      title = (f"Volume by week days"),
                      color = "forestgreen",
                      label = f"Volume" )
```

```python
plt.figure(figsize=(10,7))
plt.plot(df['Date'], df['open_close_diff'], color = "red", label = "Close - Open diff")
plt.title('Apple stock data')
plt.xlabel('Date')
plt.ylabel('Close - Open diff')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(df['Date'], df['Volume'], color = "darkblue", label = "Volume")
plt.title('Apple stock data')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```
